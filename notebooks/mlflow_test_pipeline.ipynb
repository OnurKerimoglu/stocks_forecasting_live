```python
import os

import mlflow

from raw_data_yf import fetch_ticker_data_from_yf
```

```python
ROOTPATH = os.path.dirname(os.getcwd())
MLFLOWPATH = os.path.join(ROOTPATH, "models", "mlflow_runs")
```

## Fetch data for a ticker

```python
ticker = "AMZN"
df_ticker = fetch_ticker_data_from_yf(
    ticker,
    fetchperiodinweeks=52 * 4 + 1,  # i.e., 4 years plus 1 week
)
df_ticker.head()
```

## Load exported mlflow model

```python
MODEL_ARTIFACT_FOLDER = "mlflow_models"
mlmodeldir = os.path.join(MLFLOWPATH, "test", MODEL_ARTIFACT_FOLDER)
```

```python
loaded_pipeline = mlflow.sklearn.load_model(mlmodeldir)
```

```python
# Predict on raw recent slice (must include enough lookback for lags)
y_hat = loaded_pipeline.predict(df_ticker)
y_hat
```
